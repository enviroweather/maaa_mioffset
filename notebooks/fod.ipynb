```python
import sys
import os
sys.path.append('../fod')
from mioffset.fod3 import *
from mioffset.fod3 import np
from mioffset.narr_data import read_narr_timeseries_json  # read_narr_timeseries_s3
load_dotenv()
```

starting parameters

```python
MI_LAT = 44
MI_LON = -83
ODOR_INDEX = 10
```

```python

bucket= os.getenv('NARR_BUCKET', "")
narr_index_file = os.getenv('NARR_GRID_LATLON', "")
narr_index_file_s3 = os.getenv('NARR_GRID_LATLON_S3', "")

if bucket and narr_index_file_s3:
    ts = read_narr_timeseries_json(latval = MI_LAT, 
                                   lonval= MI_LON, 
                                   bucket=bucket, 
                                   narr_grid_latlon=narr_index_file, 
                                   
                                   source = "s3")
else:
    print("Required environment variables not set for s3 retrieval")
    ts = None

type(ts)
```
